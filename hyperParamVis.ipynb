```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# let's just import our SGD progress CSV's created by gridSearch.
df1 = pd.read_csv("AnalyzedData/error_E200_L0.01.csv")
df2 = pd.read_csv("AnalyzedData/error_E200_L0.05.csv")
df3 = pd.read_csv("AnalyzedData/error_E225_L0.01.csv")
```

```python
# just some cleaning since my formatting was bad. I can go back and fix this later
df1.drop(['Unnamed: 0', '0'], axis=1,inplace=True)
df1.rename(columns={'1':'MSE'},inplace=True)
df2.drop(['Unnamed: 0', '0'], axis=1,inplace=True)
df2.rename(columns={'1':'MSE'},inplace=True)
df3.drop(['Unnamed: 0', '0'], axis=1,inplace=True)
df3.rename(columns={'1':'MSE'},inplace=True)

df2 = df2[200:].reset_index() # I have to do this to fix a mistake I made when exporting. Fine for now, 
#I fixed it in the fromScratch notebook

df3 = df3[400:].reset_index() 
```

```python
# let's visualize the MSE as a function of training epoch.
font = {'family':'serif',
       'size' : 20}

plt.figure(figsize=(15, 7))

plt.plot(df1['MSE'], label = "E = 200, LR = 0.01")
plt.plot(df2['MSE'], label = "E = 200, LR = 0.05")
plt.plot(df3['MSE'], label = "E = 225, LR = 0.01")
plt.xlabel("S.G.D. Epoch", **font)
plt.ylabel("MSE", **font)
plt.title("A plot of the MSE vs. epoch number for a few training sessions.", **font)
plt.legend()
plt.show()
```
